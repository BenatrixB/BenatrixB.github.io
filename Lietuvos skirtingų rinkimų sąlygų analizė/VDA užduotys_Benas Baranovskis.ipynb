**1 kandidatas į Lietuvos Respublikos Prezidentus būtų išrinktas pirmajame rinkimų ture**

1. Daugumos balsai: Kandidatas turi surinkti daugiau nei 50% visų galiojančių balsų. Tai reiškia, kad jis turi būti palaikomas daugiau nei pusės visų balsuojančiųjų.

2. Balsavimo dalyvių dauguma: Būtina, kad balsuotų daugiau nei pusė visų rinkėjų, kurie turėtų teisę dalyvauti rinkimuose.

3. Jis turi atitikti visas kitas reikalavimus, nurodytas Lietuvos Respublikos Konstitucijoje ir rinkimų įstatymuose, įskaitant amžių, pilietinę teisę, rezidencijos reikalavimus, neturėjimą kitų pareigų, kurios yra nesuderinamos su Prezidento pareigomis ir t.t.

"Išrinktu laikomas tas kandidatas į Respublikos Prezidentus, kuris **pirmą kartą balsuojant ir dalyvaujant ne mažiau kaip pusei visų rinkėjų, surinko daugiau kaip pusę visų rinkimuose dalyvavusių rinkėjų balsų."**

Informacija iš VRK (vyriausioji rinkimų komisija) skelbiamos rinkėjo atmintinės:
(https://www.vrk.lt/documents/10180/750415/rinkejui_apie_rinkimus.pdf/5fc22ded-170f-436c-8636-62bb6e5b7b73)

Kadangi Lietuvoje dažnai rinkėjų aktyvumas nepasiekia 50 % pirmąjame ture, organziuojamas antrasis turas, kuriame sąlygos jau kinta ir nebėra būtina surinkti 50 % aktyvumo, lieka tik tie kandidatai, kurie surinko daugiausiai balsų.

Gaunasi, kad per pirmą turą, kandidatas laimi surinkęs minimum <25% Lietuvos populiacijos balsų. Tai yra viena ketvirtoji (0,25) visų rinkėjų.

**Minimum 50% visų balsavimo teisę turinčiųjų, iš jų 50% turi atiduoti savo balsą už kandidatą.**







```python
import matplotlib.pyplot as plt

dalyvavo_rinkeju = [0, 20, 40, 60, 100]
balsai = [0, 20, 40, 60, 100]

# Reikia suskurti scatterplot daigramą, kuri rodytų pirmąsias dvi sąlygas.

# Trečioji sąlyga jau būna automatiškai patvirtinima, jei einama prie balsavimo tinkamumo sąlyga
# Nors su trečiąja sąlyga galima irgi būtų padaryti ką nors.



plt.figure(figsize=(8, 6))
plt.scatter(dalyvavo_rinkeju, balsai, marker='o', color='b', label='Kandidato Balsai')
plt.axhline(y=50, color='red', linestyle='-', label='50% riba')
plt.axvline(x=50, color='green', linestyle='-', label='50% balsavusių riba')
susikirtimas_x = 50
susikirtimas_y = 50
plt.plot([0, susikirtimas_x], [0, susikirtimas_y], color='blue', linestyle='-', linewidth=1)
plt.scatter(susikirtimas_x, susikirtimas_y, color='red', label='Susikirtimo taškas')
for i in range(len(dalyvavo_rinkeju)):
    plt.annotate(f'{balsai[i]}%', (dalyvavo_rinkeju[i] + 2, balsai[i] + 2), fontsize=10)
plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('% Dalyvavusių Rinkėjų')
plt.ylabel('% Balsų už Kandidatą')
plt.title('Prezidentas išrenkamas pirmąjame ture')
plt.legend()
plt.grid(True)
plt.show()
```

**4 Konsultaciniame referendume būtų sėkmingai pritarta referendumui pateikto sprendimo priėmimui**

Kas yra konsultacinis referendumas?



"Konsultacinis (patariamasis) referendumas laikomas įvykusiu, jeigu jame dalyvavo daugiau kaip pusė piliečių, turinčių rinkimų teisę ir įrašytų į rinkėjų sąrašą. Tuo atveju, kai konsultaciniame (patariamajame) referendume dalyvavo daugiau kaip pusė piliečių ir šiam referendumui pateikto sprendimo priėmimui pritarė ne mažiau kaip pusė jame dalyvavusių piliečių, sprendimas yra laikomas priimtu." 

Informacija iš VRK (vyriausioji rinkimų komisija) skelbiamos rinkėjo atmintinės:

(https://www.vrk.lt/documents/10180/750415/rinkejui_apie_rinkimus.pdf/5fc22ded-170f-436c-8636-62bb6e5b7b73)

Bendra Lietuvos populiacija: 2,806,000 - gyventojų 2022 metų duomenimis.

Neturi teisės maždaug 400,000 arba truputį mažiau atsižvelgiant į amžiaus grupavimą, nes amžiaus į amžiaus grupę nuo 14 iki 19 patenka ir 18-mečiai, kurie turi teisę balsuoti.
Balsavimo teisę turi maždaug 2,400,000 gyventojų 2022 metais. 

Tad "Piliečių referendumo paskelbimo iniciatyvos teisė įgyvendinama 300 000 turinčių rinkimų teisę piliečių reikalavimu."
**Reikia maždaug 12,5% visos Lietuvos turiničių balso teisę asmenų populiacijos parašų, norint teigti paraišką konsultciniam referendumui.**
Šį tašką pažymėsiu.
Remtasi šiais duomenimis.
https://osp.stat.gov.lt/statistiniu-rodikliu-analize?hash=ab672ada-e22f-409e-9ce0-33915afade8f#/

Būtina pastebėti, kad konsultacinio referendumo rezultatas neįpareigoja įstatymų leidžiamosios ar vykdančiosios valdžios kažką keisti.
Tačiau konsultacinis referendumas gali iliustruoti Lietuvos populiacijos nuomonę, kuria remiantis galima priimti sprendimus. 
Jei politikai sprendimų nepriima, rinkėjai gali jų nebeiškrinkti kituose rinkimuose ir tiek. 

**Reikia, kad minimum 50% turiničių balsavimo sąlygą dalyvautų ir 50% jų pritartų referendumui.**

```python
# Konusltacinio referendumo vienos iš sąlygų skaičiuoklė populiacijos procentais.

gyv_sk_bal = 2400000
r_salyga = 300000

sal_proc = (r_salyga * 100)/gyv_sk_bal
print(f"""Kiek procentų Lietuvos populiacijos parašų reikia, kad būtų galima teigti paraišką konsultaciniam referendumui:
      {sal_proc}%""")
```

```python
import matplotlib.pyplot as plt

dalyvavo_rinkeju = [0, 20, 40, 60, 100]
balsai = [0, 20, 40, 60, 100]

plt.figure(figsize=(8, 6))
plt.scatter(dalyvavo_rinkeju, balsai, marker='o', color='b', label='Pritariamieji balsai')
plt.axhline(y=50, color='red', linestyle='-', label='50% riba')
plt.axvline(x=50, color='green', linestyle='-', label='50% balsavusių riba')

susikirtimas_x = 50
susikirtimas_y = 50
plt.plot([0, susikirtimas_x], [0, susikirtimas_y], color='blue', linestyle='-', linewidth=1)
plt.scatter(susikirtimas_x, susikirtimas_y, color='red', label='Susikirtimo taškas')
minimum_balsai_x = 12
minimum_balsai_y = 12
plt.scatter(minimum_balsai_x, minimum_balsai_y, color='black', label='Reikalvimo sąlyga >12%')

for i in range(len(dalyvavo_rinkeju)):
    plt.annotate(f'{balsai[i]}%', (dalyvavo_rinkeju[i] + 2, balsai[i] + 2), fontsize=10)

plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('% Dalyvavusių Rinkėjų')
plt.ylabel('% Pritariamieji balsai')
plt.title('Pritarimas konsultacinio referendumo klausimui')
plt.legend()
plt.grid(True)
plt.show()
```

Reikia, kad minimum 50% visų balsavusių referendume pritrtų, kad būtų galima laikyti, jog privalomojo referendumo klausimui pritarta. Bet vis tik bendrai reikia minimum 33-34% visų balso teisę turinčiųjų populiacijos pritarimo.

(https://www.vrk.lt/documents/10180/750415/rinkejui_apie_rinkimus.pdf/5fc22ded-170f-436c-8636-62bb6e5b7b73)

Tad keletas sąlygų privalomojo referendumo vykdymui yra:
 - 12 % arba 300,000 balso teisę turinčių asmenų parašų, kad būtų galima įgyvendinti tokį referendumą.
 - 50 % balso teisę turinčių asmenų turi balsuoti, kad būtų galima laikyti referendumą įvykusiu. 
 - 50 % balsavusių arba apie 34% visos balsavimo teisę turinčios populiacijos turi pritarti.




```python
import matplotlib.pyplot as plt

dalyvavo_rinkeju = [0, 20, 40, 60, 100]
balsai = [0, 20, 40, 60, 100]

plt.figure(figsize=(8, 6))
plt.scatter(dalyvavo_rinkeju, balsai, marker='o', color='b', label='Pritariamieji balsai')
plt.axhline(y=50, color='r', linestyle='-', label='50% riba balsavusių ')
plt.axvline(x=50, color='g', linestyle='-', label='50% riba dalyvavusių')

susikirtimas_x = 50
susikirtimas_y = 50
plt.plot([0, susikirtimas_x], [0, susikirtimas_y], color='blue', linestyle='-', linewidth=1)
plt.scatter(susikirtimas_x, susikirtimas_y, color='red', label='Susikirtimo taškas')
minimum_x = 12
minimum_y = 12
plt.scatter(minimum_x, minimum_y, color='black', label='Reikalvimo sąlyga >12%')
bendra_populiacija_x = 34
bendra_populiacija_y = 34

plt.scatter(bendra_populiacija_x, bendra_populiacija_y, color='yellow', label='Bendros populiacijos minimumas 34%')
plt.axhline(y=34, color='yellow', linestyle='--')
for i in range(len(dalyvavo_rinkeju)):
    plt.annotate(f'{balsai[i]}%', (dalyvavo_rinkeju[i] + 2, balsai[i] + 2), fontsize=10)

plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('% Dalyvavusių Rinkėjų')
plt.ylabel('% Pritariamieji balsai')
plt.title('Pritarimas privalomojo referendumo klausimui')
plt.legend()
plt.grid(True)
plt.show()
```

**6 Privalomajame referendume būtų sėkmingai pritarta dėl Konstitucijos XIV skirsnio nuostatų pakeitimo**

Tai 12% sąlyga galioja. 

Ir dėl XIV skirsnio yra griežtesnės sąlygos.

"Sprendimas dėl Konstitucijos I skirsnio „Lietuvos valstybė“ ir XIV skirsnio „Konstitucijos keitimas“ nuostatų pakeitimo yra laikomas priimtu, **jeigu tam pritarė daugiau kaip pusė piliečių, turinčių rinkimų teisę ir įrašytų į rinkėjų sąrašą**"
Informacija iš VRK (vyriausioji rinkimų komisija) skelbiamos rinkėjo atmintinės:
(https://www.vrk.lt/documents/10180/750415/rinkejui_apie_rinkimus.pdf/5fc22ded-170f-436c-8636-62bb6e5b7b73)

Tai realiai turi pritarti 50% visos balsavimo teisę turinčios populiacijos.
Žinant, tai kad toks referendumas yra laikomas galiojančiu tik tada, kai 50% turinčios balsavimo teisę populiacijos balsuoja jame. Tai, jei blasuoja tik 50%, referendumas laikomas galiojančiu, bet tai dar nereiškia, kad jam yra pritarimas, nes reikia kad visi balsavę asmenys vieningai tam pritartų.
Tad čia yra 50% balsavusių ir 50% bendros populiacijos pritariančių. 


```python
import matplotlib.pyplot as plt

dalyvavo_rinkeju = [0, 20, 40, 60, 100]
balsai = [0, 20, 40, 60, 100]

plt.figure(figsize=(8, 6))
plt.scatter(dalyvavo_rinkeju, balsai, marker='o', color='b', label='Pritariamieji balsai')
plt.axhline(y=50, color='r', linestyle='-', label='50% riba balsavusių ')
plt.axvline(x=50, color='g', linestyle='-', label='50% riba dalyvavusių')

susikirtimas_x = 50
susikirtimas_y = 50
plt.plot([0, susikirtimas_x], [0, susikirtimas_y], color='blue', linestyle='-', linewidth=1)
plt.scatter(susikirtimas_x, susikirtimas_y, color='red', label='Susikirtimo taškas')
minimum_x = 12
minimum_y = 12
plt.scatter(minimum_x, minimum_y, color='black', label='Reikalvimo sąlyga >12%')
bendra_populiacija_x = 50
bendra_populiacija_y = 50
plt.scatter(bendra_populiacija_x, bendra_populiacija_y, color='yellow', label='Bendros populiacijos minimumas 50%')

for i in range(len(dalyvavo_rinkeju)):
    plt.annotate(f'{balsai[i]}%', (dalyvavo_rinkeju[i] + 2, balsai[i] + 2), fontsize=10)

plt.xlim(0, 110)
plt.ylim(0, 110)
plt.xlabel('% Dalyvavusių Rinkėjų')
plt.ylabel('% Pritariamieji balsai')
plt.title('Pritarimas privalomojo referendumo klausimui dėl Konstitucijos XIV skirsnio pakeitimų')
plt.legend()
plt.grid(True)
plt.show()
```
