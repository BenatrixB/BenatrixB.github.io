# **1. Kuri marketingo kampanijos banga pritraukė daugiausiai žmonių?**

```python
# kodas pasibandymams
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
m_data = pd.read_csv('marketing_data.csv')

import pandas as pd
import matplotlib.pyplot as plt

kampanijos = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5"]
rezultatai = m_data[kampanijos].sum()

rezultatai.plot(kind='bar', color = 'black', edgecolor='lightgreen', xlabel='Banga', ylabel='Žmonių skaičius', title='Kiek žmonių priėmė kampaniją')
plt.show()
```

Galime matyti, kad sėkmingiausia buvo 4 banga kampanijos, kuri pritraukė 167 asmenis.

```python
# Rasti min ir max šitame bar charte
bangos = m_data[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']]

atsakymu_sumos = bangos.sum()

# Randame mažiausią ir didžiausią bendrą atsakymų skaičių kiekvienoje bangos grupėje
maziausia_suma = atsakymu_sumos.min()
didziausia_suma = atsakymu_sumos.max()

print("Bendros atsakymų sumos:")
print(atsakymu_sumos)

print("\nMažiausia bendra atsakymų suma:", maziausia_suma)
print("Didžiausia bendra atsakymų suma:", didziausia_suma)
```

# **2. Rasti skirtingų pardavimų prieigų vidurkius**

```python
web_vidurkis = round(m_data['NumWebPurchases'].mean())
print(f"Kiek vartotojai vidutiniškai padarė pirkimų naudojant web parduotuvę: {web_vidurkis}")
cat_vidurkis = round(m_data['NumCatalogPurchases'].mean())
print(f"Kiek vartotojai vidutiniškai padarė pirkimų naudojant katalogą: {cat_vidurkis}")
store_vidurkis = round(m_data['NumStorePurchases'].mean())
print(f"Kiek vartotojai vidutiniškai padarė pirkimų parduotuvėje: {store_vidurkis}")

# Ieškoma didžiausio vidurkio
if web_vidurkis > cat_vidurkis and web_vidurkis > store_vidurkis:
    print(f"Web parduotuvės vidurkis didžiausias: {web_vidurkis}")
elif cat_vidurkis > web_vidurkis and cat_vidurkis > store_vidurkis:
    print(f"Katalogo pardavimai turi didžiausią vidurkį:{cat_vidurkis}")
elif store_vidurkis > web_vidurkis and store_vidurkis > cat_vidurkis:
    print(f"Parduotuvių pardavimai turi didžiausią vidurkį: {store_vidurkis}")
```

# **5. Kurioje šalyje yra daugiausiai klientų?**

```python

klientu_skaiciai = m_data['Country'].value_counts()
daugiausiai_klientu_salis = klientu_skaiciai.idxmax()
klientu_skaicius = klientu_skaiciai.max()
# bar chart
plt.figure(figsize=(10, 6))
klientu_skaiciai.plot(kind='bar', color='black', edgecolor='lightgreen')
plt.title('Klientų skaičius pagal šalis')
plt.xlabel('Šalis')
plt.ylabel('Klientų skaičius')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
print()
print(f"Daugiausiai klientų yra šalyje {daugiausiai_klientu_salis} su {klientu_skaicius} klientais.")





```
